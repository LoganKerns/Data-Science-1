```python
from bs4 import BeautifulSoup
import requests
import re

# Define the URL of the Wikipedia article
wikipedia_url = "https://en.wikipedia.org/wiki/Technological_singularity"

# Set the user agent in the request headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}

# Send a GET request with headers
response = requests.get(wikipedia_url, headers=headers)

# Parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find all paragraph elements within the content
paragraphs = soup.find_all("p")

# Extract and concatenate the text from all paragraphs
article_contents = "\n".join([p.get_text() for p in paragraphs])

# Split the text into words using regular expressions
words = re.findall(r'\b\w+\b', article_contents)

# Count the number of words
num_words = len(words)

# Print the number of words
print(f"The article contains {num_words} words.")
```
