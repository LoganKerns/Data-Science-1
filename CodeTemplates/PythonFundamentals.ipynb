```python
# Challenge 1
df = pd.read_csv("pirates.csv") # This requires the csv file to be in the same folder as this Jupyter notebook.
```

```python
# Challenge 2 - Part 3
for column in df.columns:
    for row in range(len(df["id"])):
        if pd.isna(df.loc[row, column]):
            print(row + 1)
            print(column)
```

```python
# Challenge 3
# Technique 1
# This drops all rows that contain a null value

df.dropna()
```

```python
# Challenge 3
# Technique 2
# Drops just the age row that features a null

df.drop(33, inplace=True)
```

```python
# Challenge 3
# Technique 3
# Reassigns the missing value in age to the mean of ages

df.loc[33, "age"] = df["age"].mean()
```

```python
# Challenge 3
# Technique 4
# Reassigns the missing value in favorite.pirate to the mode

df.loc[634, "favorite.pirate"] = df["favorite.pirate"].mode()[0]
```

```python
# Challenge 3
# Technique 5
# Assigns the missing age value to a random integer in the range of the smallest age in the dataset and the largest age in dataset
df.loc[33, "age"] = random.randint(df["age"].min(), df["age"].max())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('compas.csv')

age_groups = ['Younger than 25', 'Between 25 and 45', 'Older than 45']

age_cutoffs = [0, 25, 45, float('inf')]

data['Age Group'] = pd.cut(data['age'], bins=age_cutoffs, labels=age_groups)

age_group_counts = data['Age Group'].value_counts()

for age_group, count in age_group_counts.items():
    print(f'{age_group}: {count} people')

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6)) 
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette="pastel")

plt.title('Age Distribution in the Compas Dataset')
plt.xlabel('Age Group')
plt.ylabel('Count')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (compas.csv)
data = pd.read_csv('compas.csv')

# Convert 'c_jail_in' and 'c_jail_out' to datetime objects
data['c_jail_in'] = pd.to_datetime(data['c_jail_in'])
data['c_jail_out'] = pd.to_datetime(data['c_jail_out'])

# Calculate 'days_in_jail' by subtracting 'c_jail_in' from 'c_jail_out'
data['days_in_jail'] = (data['c_jail_out'] - data['c_jail_in']).dt.days

# Set the style of the plot
sns.set(style="whitegrid")

# Create the scatterplot
plt.figure(figsize=(8, 6))  # Set the figure size
sns.scatterplot(data=data, x='days_in_jail', y='age', color='skyblue', alpha=0.5)

# Customize the plot
plt.title('Scatterplot of Days in Jail vs. Age')
plt.xlabel('Days in Jail')
plt.ylabel('Age')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (compas.csv)
data = pd.read_csv('compas.csv')

# Count the number of individuals in each race category
race_counts = data['race'].value_counts()

# Set the style of the plot
sns.set(style="whitegrid")

# Create the bar graph
plt.figure(figsize=(10, 6))  # Set the figure size
barplot = sns.barplot(x=race_counts.index, y=race_counts.values, palette="pastel")

# Customize the plot
plt.title('Race Distribution in the Compas Dataset')
plt.xlabel('Race')
plt.ylabel('Count')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Add annotations (count labels) on top of the bars
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     fontsize=12, color='black', xytext = (0, 5),
                     textcoords = 'offset points')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (compas.csv)
data = pd.read_csv('compas.csv')

# Create a histogram with counts
plt.figure(figsize=(10, 6))  # Set the figure size

# Create the histogram and capture the bin counts
n, bins, _ = plt.hist(data['v_decile_score'], bins=10, color='black', alpha=0.7)

# Customize the plot
plt.title('Histogram of v_decile_score in the Compas Dataset')
plt.xlabel('v_decile_score')
plt.ylabel('Count')

# Add count labels on top of each bar
for count, x in zip(n, bins):
    if count > 0:
        plt.text(x + 0.5, count, str(int(count)), ha='center', va='bottom', fontsize=10)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (compas.csv)
data = pd.read_csv('compas.csv')

# Create separate dataframes for male and female
male_data = data[data['sex'] == 'Male']
female_data = data[data['sex'] == 'Female']

# Create histograms with counts for males and females
plt.figure(figsize=(12, 6))  # Set the figure size

# Create the histogram for males
n, bins, _ = plt.hist(male_data['v_decile_score'], bins=10, alpha=0.5, label='Male', color='blue')

# Customize the plot
plt.title('Histogram of v_decile_score by Gender (Male)')
plt.xlabel('v_decile_score')
plt.ylabel('Count')

# Add count labels on top of each bar for males
for count, x in zip(n, bins):
    if count > 0:
        plt.text(x + 0.5, count, str(int(count)), ha='center', va='bottom', fontsize=10)

# Create the histogram for females
n, bins, _ = plt.hist(female_data['v_decile_score'], bins=10, alpha=0.5, label='Female', color='purple')

# Customize the plot for females
plt.title('Histogram of v_decile_score by Gender')
plt.xlabel('v_decile_score')
plt.ylabel('Count')

# Add count labels on top of each bar for females
for count, x in zip(n, bins):
    if count > 0:
        plt.text(x + 0.5, count, str(int(count)), ha='center', va='bottom', fontsize=10)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
