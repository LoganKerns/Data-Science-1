```python
import requests

# Replace 'YOUR_API_KEY' with your actual TMDb API key
api_key = '6fa4f5b97c2e25123f294d1c66481725'

# Define the base URL for TMDb API
base_url = 'https://api.themoviedb.org/3'

# Create a function to fetch the highest-rated movies of a decade from multiple pages
def get_highest_rated_movies_of_decade(start_year, end_year, max_pages=5):
    highest_rated_movies = []

    for page in range(1, max_pages + 1):
        # Define query parameters
        params = {
            'api_key': api_key,
            'primary_release_date.gte': f'{start_year}-01-01',
            'primary_release_date.lte': f'{end_year}-12-31',
            'sort_by': 'vote_average.desc',  # Sort by vote average in descending order
            'page': page  # Get the current page of results
        }

        # Make the API request to discover movies in the specified decade
        response = requests.get(f'{base_url}/discover/movie', params=params)

        if response.status_code == 200:
            movie_data = response.json()
            if movie_data['results']:
                # Append the highest-rated movie from the current page to the list
                highest_rated_movies.append(movie_data['results'][0])
            else:
                break  # No more results on subsequent pages
        else:
            print(f'Error: {response.status_code}')
            break  # Error occurred, stop fetching more pages

    return highest_rated_movies

# Create a list to store the highest-rated movies of each decade
highest_rated_movies = []

# Loop through each decade from 1920 to the present
for start_year in range(1920, 2030, 10):
    end_year = start_year + 9
    highest_rated_movies.extend(get_highest_rated_movies_of_decade(start_year, end_year))

# Sort the list of highest-rated movies by vote_average in descending order
highest_rated_movies.sort(key=lambda x: x['vote_average'], reverse=True)

# Display the highest-rated movies of each decade
for movie in highest_rated_movies:
    print(f'Decade: {movie["release_date"][:4]}s')
    print(f'Title: {movie["title"]}')
    print(f'Vote Average: {movie["vote_average"]}')
    print('---')
```
